# Data Augmentation

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img

import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('E:/CVPR/Dataset/Deb Sir Augmentation Test/Hernia/00000003_001.png')
#Converts from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize the image to (300, 300)
img_resized = cv2.resize(img, (300, 300))

plt.imshow(img)
plt.show()
```

```python
print(img.shape)

img_reshape = img_resized.reshape( 1, 300, 300, 3) # 1, 300, 300, 3
print(img_reshape.shape)
```

```python
img2 = img_resized
plt.imshow(img2)
plt.show()
```

```python
datagen = ImageDataGenerator(
    rotation_range=15,  # Small rotations
    width_shift_range=0.1,  # Small horizontal shifts
    height_shift_range=0.1,  # Small vertical shifts
    shear_range=0.1,  # Minimal shear
    zoom_range=0.1,  # Small zoom
    horizontal_flip=False,  # No horizontal flip (use with caution if set to True)
    vertical_flip=False,  # No vertical flip
    fill_mode='nearest'  # Use nearest fill mode
)
```

```python
i = 0
for batch in datagen.flow(img_reshape, 
                          save_to_dir='E:/CVPR/Dataset/Deb Sir Augmentation Test/HerniaAugResult',
                          save_prefix='aug_',
                          save_format='jpg'
                         ):
    i += 1
    if i >= 31:
        break
```

```python
i = 0
for batch in datagen.flow_from_directory(directory='D:/Edema', 
                          batch_size=4,
                          save_to_dir='D:/he/Pneumonia_700',
                          save_prefix='aug_',
                      save_format='jpg'
                         ):
    i += 1
    if i >=1:
        break
```
